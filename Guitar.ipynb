# Guitar Zero

```python

note=['a','a#','b','c','c#','d','d#','e','f','f#','g','g#','a','a#','b','c','c#','d','d#','e','f','f#','g','g#']
str1_e=note[note.index('e'):note.index('e')+12]
str2_b=note[note.index('b'):note.index('b')+12]
str3_g=note[note.index('g'):note.index('g')+12]
str4_d=note[note.index('d'):note.index('d')+12]
str5_a=note[note.index('a'):note.index('a')+12]
strings=[str1_e,str2_b,str3_g,str4_d,str5_a,str1_e]
```

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
for i,string in enumerate(strings):
    x = i
    print(i,string)
```

```python
for i,string in enumerate(strings):
    x = i
    print(i,end=' -> ')
    for j, note in enumerate(string):
        print(j,end='')
    print()
```

```python
for x,string in enumerate(strings):
    for y, note in enumerate(string):
        plt.scatter(x, y)
        plt.text(x+0.3, y+0.3, note, fontsize=9)
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(16,5))

for x,string in enumerate(strings):
    for y, note in enumerate(string):
        plt.text(y+.5, x+2,note, fontsize=12)
plt.grid(True)
plt.axis([0, 15, 0, 9])
labels = np.arange(1,16,1)
plt.xticks(labels, labels)
ylabels=['.','E','B','g','D','A','E','.','.']
ynum=np.arange(0,9,1)

plt.yticks(ynum,ylabels)
plt.show()
```
